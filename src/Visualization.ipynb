```python
import pickle
import matplotlib.pyplot as plt
import cairosvg
import matplotlib.image as mpimg
%matplotlib inline
```

```python
with open('training_results.pkl', 'rb') as f:
    train_losses, eval_losses, train_corrects, eval_corrects = pickle.load(f)

train_accuracies = [correct / 60000 for correct in train_corrects]
eval_accuracies = [correct / 10000 for correct in eval_corrects]
```

```python
# Convert SVG to PNG
cairosvg.svg2png(url="nn.svg", write_to="converted_image.png")
img = mpimg.imread("converted_image.png")
plt.figure(figsize=(20,10))
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(train_losses, label="Training Loss", color='blue')
plt.plot(eval_losses, label="Validation Loss", color='orange')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title("Training and Validation Loss over Epochs")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(train_accuracies, label='Training Accuracy', marker='o', color='blue')
plt.plot(eval_accuracies, label='Validation Accuracy', marker='o', color='orange')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy Over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```
